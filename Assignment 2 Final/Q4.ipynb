### Estimate Gaussian parameters

Write a function to estimate the parameters of the Gaussian distribution.

```python
# your code here
def dataproperties(data):
    mean=np.mean(data)
    std=np.std(data)
    return mean,std

def gaussian(x,mean,std):
    return (1/(std*np.sqrt(2*np.pi)))*np.exp(-((x-mean)**2)/(2*std**2))
```

### Load Data

The data has been loaded onto a Pandas DataFrame. Try to get an initial feel for the data by using functions like `describe()`, `info()`, or maybe try to plot the data to check for any patterns.

```python
data = pd.read_csv('wifiLocalization.txt', sep='\t')
data.describe()

X = data.iloc[:, 0:7].values
y = data.iloc[:, 7].values
```

### Splitting the Data

It is a good practice to split the data into training and test sets. This is to ensure that the model is not overfitting to the training data. The test set is used to evaluate the performance of the model on unseen data. The test set is not used to train the model in any way.
```python
# your code here
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

### Calculate priors

Write a function to calculate the priors for each class.

```python
prior = np.zeros(4)
for i in range(4):
    prior[i] = np.sum(y_train == i+1)/len(y_train)

prior
    
```

```python
#calculating mean and std for each class

classproperties=np.zeros((4,7,2)) 

for i in range(4):
    list=[]
    for j in range(7):
        class_mean,class_std=dataproperties(X_train[y_train==i+1,j])
        list.append([class_mean,class_std])
    classproperties[i]=list

print(classproperties)
```

### Likelihood + Classification

Given a test sample, write a function to get the likelihoods for each class in the sample. Use the Gaussian parameters and priors calculated above. Then compute the likelihood that the sample belongs to each class and return the class with the highest likelihood.

What is a common problem with the likelihoods? How can you fix it? Redo the classification with the fixed likelihoods. (You can either write another function or modify the existing one after mentioning the reason for the change)

```python
# your code here

def likelihoods(data,classproperties,attributecount=7):
    predictclass=[]
    for i in data:
        scores=[]
        for j in range(4):
            likelihood=0
            for k in range(attributecount):
                likelihood+=np.log(gaussian(i[k],classproperties[j,k,0],classproperties[j,k,1]))
            likelihood+=np.log(prior[j])
            scores.append(likelihood)
        predictclass.append(np.argmax(scores)+1)
    return predictclass
```

### 1) Estimating model on training data

```python
# your code here
predicted=likelihoods(X_train,classproperties,7)

print("Accuracy:",(np.sum(predicted==y_train)/len(predicted))*100)
```

### 2) Plotting gaussian probability density functions for each class after estimation

```python
for i in range(4):
    plt.figure(figsize=(16,4))
    for j in range(7):
        x=np.sort(X_train[y_train==i+1,j])
        plt.plot(x,gaussian(x,classproperties[i,j,0],classproperties[i,j,1]),label="Attribute "+str(j+1))
    plt.title("For Room "+str(i+1))

    plt.legend()
    plt.show()
```

```python
predicted=likelihoods(X_test,classproperties,7)

print("Accuracy:",(np.sum(predicted==y_test)/len(predicted))*100)
```

### 4) Plotting misclassified samples

```python
misclassified = np.where(predicted != y_test)[0]

tracker=1

for i in misclassified:
    print("Misclassified sample "+str(tracker)+":")
    tracker+=1
    print("Actual class:",y_test[i])
    print("Predicted class:",predicted[i])
    for i in range(4):
        plt.figure(figsize=(16,4))
        for j in range(7):
            x=np.sort(X_train[y_train==i+1,j])
            plt.plot(x,gaussian(x,classproperties[i,j,0],classproperties[i,j,1]),label="Attribute "+str(j+1))
            plt.plot(X_test[i,j],gaussian(X_test[i,j],classproperties[i,j,0],classproperties[i,j,1]),'ro')
        plt.title("For Room "+str(i+1))
        plt.legend()
        plt.show()
```

### 5) Identifying redundant features

```python
#obtaining accuracies my excluding each attribute one by one
for i in range(7):
    X_train1=np.delete(X_train,i,1)
    X_test1=np.delete(X_test,i,1)
    classproperties1=np.zeros((4,6,2)) 

    for j in range(4):
        list=[]
        for k in range(6):
            class_mean,class_std=dataproperties(X_train1[y_train==j+1,k])
            list.append([class_mean,class_std])
        classproperties1[j]=list
    predicted1=likelihoods(X_test1,classproperties1,6)
    print("Accuracy excluding attribute "+str(i+1)+":",(np.sum(predicted1==y_test)/len(predicted1))*100)
    
    
    
```

### 6) Identifying overpowering features

```python
# identifying accuracies by taking each attribute one by one
for i in range(7):
    X_train1=X_train[:,i].reshape(-1,1)
    X_test1=X_test[:,i].reshape(-1,1)
    classproperties1=np.zeros((4,1,2))
    
    for j in range(4):
        class_mean,class_std=dataproperties(X_train1[y_train==j+1])
        classproperties1[j]=[[class_mean,class_std]]
    predicted1=likelihoods(X_test1,classproperties1,1)
    print("Accuracy for attribute "+str(i+1)+":",(np.sum(predicted1==y_test)/len(predicted1))*100)
    
```

### 7) Using sklearn

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

classifier = GaussianNB()
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print("Accuracy:",accuracy_score(y_test, y_pred)*100)
```

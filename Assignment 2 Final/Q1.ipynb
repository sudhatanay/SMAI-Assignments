## Part1: K Means

```python
from keras.datasets import fashion_mnist
(trainX, trainy), (testX, testy) = fashion_mnist.load_data()

trainX = trainX.astype('float32') 
testX = testX.astype('float32')

trainX=trainX/255.0
testX=testX/255.0

final_trainX = trainX.reshape(len(trainX),-1)
final_testX = testX.reshape(len(testX),-1)
# # you can performs your training on train/test dataset alone if there are computational issues
```

#### Training was performed on test dataset alone

```python
#implement elbow method from scratch with kmeans++
def elbow(X):
    avg_dist = []
    K = [10,20,30,40,50]

    for k in tqdm(K):
    # Building and fitting the model
        kmeanModel = KMeans(n_clusters=k,n_init=10,init='k-means++')
        kmeanModel.fit(X)

        avg_dist.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,'euclidean'), axis=1)) / X.shape[0])
        

    plt.plot(K, avg_dist, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel('Avg Dist')
    plt.title('Elbow')
    plt.show()
    
elbow(final_testX)
```

```python
#implement elbow method from scratch with random
def elbow(X):
    avg_dist = []
    K = [10,20,30,40,50]

    for k in tqdm(K):
    # Building and fitting the model
        kmeanModel = KMeans(n_clusters=k,n_init=10,init='random')
        kmeanModel.fit(X)

        avg_dist.append(sum(np.min(cdist(X, kmeanModel.cluster_centers_,'euclidean'), axis=1)) / X.shape[0])
        

    plt.plot(K, avg_dist, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel('Avg Dist')
    plt.title('Elbow')
    plt.show()
    
elbow(final_testX)
```

```python
#implement silhouette method from scratch with kmeans++
def silhouette(X):
    silhouette_scores = []
    K = [10,20,30,40,50]

    for k in tqdm(K):
    # Building and fitting the model
        kmeanModel = KMeans(n_clusters=k,n_init=10,init='k-means++')
        
        clusters = kmeanModel.fit_predict(X)
        silhouette_scores.append(silhouette_score(X, clusters))
        
    plt.plot(K, silhouette_scores, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel('Silhouette scores')
    plt.title('Silhouette')
    plt.show()
    
silhouette(final_testX)
```

```python
#implement silhouette method from scratch with random
def silhouette(X):
    silhouette_scores = []
    K = [10,20,30,40,50]

    for k in tqdm(K):
    # Building and fitting the model
        kmeanModel = KMeans(n_clusters=k,n_init=10,init='random')
        
        clusters = kmeanModel.fit_predict(X)
        silhouette_scores.append(silhouette_score(X, clusters))
        
    plt.plot(K, silhouette_scores, 'bx-')
    plt.xlabel('Values of K')
    plt.ylabel('Silhouette scores')
    plt.title('Silhouette')
    plt.show()
    
silhouette(final_testX)
```

```python
#visualize a dendogram

def dendogram(X,meth):
    plt.figure(figsize=(10,6))
    dend = sch.dendrogram(sch.linkage(X, method = meth))
    plt.title('Dendrogram')
    plt.xlabel('Samples')
    plt.ylabel('Euclidean distances')
    plt.show()
    
```

```python
# implement hierarchical clustering ( you can use scikit module)
chlabels=[]
shlabels=[]

def hierarchical_clustering(X,threshholdval,meth):
    clustermodel= AgglomerativeClustering(n_clusters=None, metric=None, linkage=meth, distance_threshold=threshholdval).fit_predict(X)
    return clustermodel

chlabels = hierarchical_clustering(chtestX,12.5,'complete')
shlabels = hierarchical_clustering(shtestX,4,'single')
```

**Question**: Explain the procedure to find the optimal number of clusters using dendogram

Answer: We can use the largest vertical distance in our dendrogram and draw a line between them to find the optimal number of clusters. Else, we can set a cutoff or distance threshhold and form clusters based on that as well

**Question:** Mention the differences observed using single and complete linkage:

Answer:
Single linkage provides clusters that are more spread out as compared to complete linkage clusters. Single linkage computes cluster distance based on the closest two points between clusters while complete linkage uses the farthest two points between clusters

## Part3: Image Compression

```python
# Boiler plate code
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_size = img.shape
rows=img_size[0]
cols=img_size[1]

# Reshape it to be 2-dimension
X = img.reshape(img_size[0] * img_size[1], img_size[2])
```

```python
count=0
fig = plt.figure(figsize=(10,5))

for iterable,i in enumerate(range(5,105,10),start=1):
    kmeans = MiniBatchKMeans(n_clusters=i,n_init=10)
    kmeans.fit(X)
    
    compressed_X = kmeans.cluster_centers_[kmeans.labels_]
    compressed_X = np.clip(compressed_X.astype('uint8'), 0, 255)
    compressed_X = compressed_X.reshape(rows, cols, 3)
    
    
    k=fig.add_subplot(2, 5, iterable)
    title = "K = " + str(i)
    k.title.set_text(title)
    plt.imshow(compressed_X)
    plt.axis('off')

    
```

We can notice the compression works as we see the graining near the lightsource get more pronounced as K value decreases
